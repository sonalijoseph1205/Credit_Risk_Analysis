Data handling and cleaning 

```python
inp1.drop('Unnamed: 0',axis=1, inplace= True)
```

DAYS_BIRTH column

```python
inp0.DAYS_BIRTH.head()
```

 Age is given in days with a "-" in front of it so that will need cleaning:
 
 a)Removing -
 
 b)Converting Age in days to Age in years

```python
inp0["Age"]=inp0["DAYS_BIRTH"].apply(lambda x: x//365 if x>0 else abs(x//365))
```

```python
inp0.Age.describe()
```

Binning age into groups

```python
inp0["age_bins"]=pd.cut(x=inp0.Age,bins=[20,30,40,50,60,999],labels=["20-30","30-40","40-50","50-60","60+"])
inp0.head()
```

Handling the Education column

```python
inp0.NAME_EDUCATION_TYPE.head(10)
```

Removing the data after "/"

```python
inp0["education"]=inp0.NAME_EDUCATION_TYPE.apply(lambda x : str(x.split("/")[0]))
```

```python
inp0.education.describe()
```

```python
inp0.education.head()
```

Checking the DAYS_EMPLOYED column:

```python
inp0.DAYS_EMPLOYED.head()
```

Removing "-" 

```python
inp0["DAYS_EMPLOYED"]=inp0.DAYS_EMPLOYED.apply(lambda x: abs(x))
```

```python
inp0.DAYS_EMPLOYED.dtype
```
